```python
import pandas as pd
import numpy as np
import os
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Load the dataset
df = pd.read_csv('/content/ML-EdgeIIoT-dataset.csv', low_memory=False)

# Encode categorical features
label_col = 'Attack_label'
for col in df.select_dtypes(include='object'):
    if col != label_col:
        df[col] = LabelEncoder().fit_transform(df[col])
df[label_col] = LabelEncoder().fit_transform(df[label_col])

# Standardize numeric features
X = df.drop(columns=[label_col])
y = df[label_col].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Non-IID Split using Dirichlet distribution
def non_iid_split(X, y, num_clients=5, alpha=0.5):
    classes = np.unique(y)
    idx_by_class = {c: np.where(y == c)[0] for c in classes}
    client_data = {i: [] for i in range(num_clients)}

    for c in classes:
        idx = idx_by_class[c]
        np.random.shuffle(idx)
        proportions = np.random.dirichlet(np.repeat(alpha, num_clients))
        proportions = (proportions * len(idx)).astype(int)
        proportions[-1] = len(idx) - proportions[:-1].sum()  # fix rounding
        start = 0
        for i, count in enumerate(proportions):
            client_data[i].extend(idx[start:start + count])
            start += count

    return client_data

# Generate and save client datasets
client_indices = non_iid_split(X_scaled, y, num_clients=5, alpha=0.5)
os.makedirs("clients", exist_ok=True)

for i, idx in client_indices.items():
    Xi, yi = X_scaled[idx], y[idx]
    np.savez(f"clients/client_{i}.npz", X=Xi, y=yi)
    print(f"Client {i}: {len(idx)} samples")
```

```python
import flwr as fl
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import os

# -------------- Load the Split Client Data ---------------- #
CLIENTS = 5
client_data = []
for i in range(CLIENTS):
    data = np.load(f"clients/client_{i}.npz")
    X_train, X_test, y_train, y_test = train_test_split(
        data["X"], data["y"], test_size=0.2, random_state=42)
    client_data.append((X_train, y_train, X_test, y_test))

# -------------- Define the Federated Client ---------------- #
class FederatedClient(fl.client.NumPyClient):
    def __init__(self, X_train, y_train, X_test, y_test):
        self.model = LogisticRegression(max_iter=200)
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test

    def get_parameters(self, config):
        return [self.model.coef_, self.model.intercept_]

    def fit(self, parameters, config):
        self.model.coef_, self.model.intercept_ = parameters
        self.model.fit(self.X_train, self.y_train)
        return self.get_parameters(config), len(self.X_train), {}

    def evaluate(self, parameters, config):
        self.model.coef_, self.model.intercept_ = parameters
        preds = self.model.predict(self.X_test)
        acc = accuracy_score(self.y_test, preds)
        return float(acc), len(self.X_test), {}

# -------------- Start Federated Simulation ---------------- #
def client_fn(cid: str):
    i = int(cid)
    X_train, y_train, X_test, y_test = client_data[i]
    return FederatedClient(X_train, y_train, X_test, y_test)

fl.simulation.start_simulation(
    client_fn=client_fn,
    num_clients=CLIENTS,
    config=fl.server.ServerConfig(num_rounds=5),
)
```

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import pickle

# -------------------------------
# 1. Load and preprocess dataset
# -------------------------------
# Adjust path if needed
df = pd.read_csv("/content/ML-EdgeIIoT-dataset.csv", low_memory=False) # Use low_memory=False to avoid DtypeWarning

# Separate target variable
label_col = 'Attack_label' # Assuming 'Attack_label' is the target column
if label_col not in df.columns:
    label_col = df.columns[-1] # Fallback to last column if 'Attack_label' is not found

y = df[label_col]
X = df.drop(columns=[label_col])

# Encode attack labels
attack_label_encoder = LabelEncoder()
y_encoded = attack_label_encoder.fit_transform(y)

# Separate numerical and categorical features
num_cols = X.select_dtypes(include=np.number).columns
cat_cols = X.select_dtypes(include='object').columns

# Encode categorical features
encoders = {}
for col in cat_cols:
    le = LabelEncoder()
    # Convert to string and handle potential NaNs before encoding
    X[col] = X[col].astype(str)
    X[col] = le.fit_transform(X[col])
    encoders[col] = le

# Now all features should be numerical or encoded numerical.
# Proceed with train-test split on the processed data
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, stratify=y_encoded, random_state=42
)


# Scale numerical features (select only numeric columns from the split dataframes)
# Ensure only numeric columns are passed to the scaler after encoding categorical
X_train_num = X_train[num_cols]
X_test_num = X_test[num_cols]

scaler = StandardScaler()
X_train_scaled_num = scaler.fit_transform(X_train_num)
X_test_scaled_num = scaler.transform(X_test_num)

# Combine scaled numerical features with encoded categorical features if needed,
# but for this model architecture, all features should be numerical.
# Assuming all columns in X are intended to be numerical after encoding:
X_train_scaled = X_train_scaled_num
X_test_scaled = X_test_scaled_num

# Reshape for CNN+LSTM (samples, timesteps, features)
X_train_scaled = X_train_scaled.reshape((X_train_scaled.shape[0], 1, X_train_scaled.shape[1]))
X_test_scaled = X_test_scaled.reshape((X_test_scaled.shape[0], 1, X_test_scaled.shape[1]))

num_classes = len(np.unique(y_encoded))
input_shape = (X_train_scaled.shape[1], X_train_scaled.shape[2])  # (timesteps, features)

# -------------------------------
# 2. Define Hybrid CNN+LSTM model
# -------------------------------
def build_hybrid_cnn_lstm(input_shape, num_classes):
    model = keras.Sequential()
    # CNN layer to extract spatial patterns
    model.add(layers.Conv1D(filters=64, kernel_size=3, activation="relu", padding="same", input_shape=input_shape))
    # Removed MaxPooling1D as it's not compatible with timestep=1
    # model.add(layers.MaxPooling1D(pool_size=2))
    # LSTM layer to capture temporal dependencies
    model.add(layers.LSTM(64, return_sequences=False))
    model.add(layers.Dropout(0.5))
    # Fully connected
    model.add(layers.Dense(64, activation="relu"))
    model.add(layers.Dense(num_classes, activation="softmax"))

    model.compile(
        optimizer="adam",
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"]
    )
    return model

hybrid_cnn_lstm_model = build_hybrid_cnn_lstm(input_shape, num_classes)

# -------------------------------
# 3. Train model
# -------------------------------
history = hybrid_cnn_lstm_model.fit(
    X_train_scaled, y_train,
    validation_data=(X_test_scaled, y_test),
    epochs=10, batch_size=64, verbose=1
)

# -------------------------------
# 4. Save artifacts
# -------------------------------
# Save scaled training data
np.save("X_train_scaled.npy", X_train_scaled)

# Save scaler
with open("scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

# Save label encoders (the dictionary of fitted encoders)
with open("encoders.pkl", "wb") as f:
    pickle.dump(encoders, f)


# Save trained model
hybrid_cnn_lstm_model.save("hybrid_cnn_lstm_multiclass.h5")

print("✅ Saved: X_train_scaled.npy, scaler.pkl, encoders.pkl, hybrid_cnn_lstm_multiclass.h5")

# -------------------------------
# 5. (Optional) Load back later
# -------------------------------
# Load scaler
# with open("scaler.pkl", "rb") as f:
#     scaler_loaded = pickle.load(f)

# Load label encoders
# with open("encoders.pkl", "rb") as f:
#     encoders_loaded = pickle.load(f)

# Load model
# model_loaded = tf.keras.models.load_model("hybrid_cnn_lstm_multiclass.h5")
```
